```python
import requests
import numpy as np
from bs4 import BeautifulSoup
import pandas as pd
import json
```

```python
def Cars_Scrapping():
    title = []
    prices = []
    model = []
    Km_Driven = []
    Engine_type = []
    CC = []
    Transmission = []

    for i in range(1, 2132):
        url = f'https://www.pakwheels.com/used-cars/search/-/?page={i}'
        response_car = requests.get(url)
        soup_car = BeautifulSoup(response_car.text, 'html.parser')
        
        all_divs = soup_car.find_all('div', {'class': 'col-md-9 grid-style'})
        
        for all_div in all_divs:
            try:
                price = all_div.find('div', {'class': 'price-details generic-dark-grey'}).text.strip()
                if 'corer' in price:
                    price = price * 100
                price = price.replace('PKR', '').replace('lacs', '').replace('crore', '').replace('Call', '').strip()

                if price:
                    prices.append(price)
                else:
                    prices.append(np.nan)
                
            except AttributeError:
                price = None
        
            all_h3_tags = all_div.find_all('h3')
            
            for h3_tag in all_h3_tags:
                titles = h3_tag.text.strip()
                if titles:
                    title.append(titles)
                else:
                    title.append(np.nan)

                t = all_div.find_all('li')

                models = t[1].text.strip()
                if models:
                    model.append(models)
                else:
                    model.append(np.nan)

                Km_Drivens = t[2].text.strip()
                if Km_Drivens:
                    Km_Driven.append(Km_Drivens)
                else:
                    Km_Driven.append(np.nan)

                Engine_types = t[3].text.strip()
                if Engine_types:
                    Engine_type.append(Engine_types)
                else:
                    Engine_type.append(np.nan)
                        
                CCs = t[4].text.strip()
                if CCs:
                    CC.append(CCs)
                else:
                    CC.append(np.nan)
                        
                Transmissions = t[5].text.strip()
                if Transmissions:
                    Transmission.append(Transmissions)
                else:
                    Transmission.append(np.nan)
                
    return title, prices, model, Km_Driven, Engine_type, CC, Transmission
    
title, prices, model, Km_Driven, Engine_type, CC, Transmission = Cars_Scrapping()

Data = {
    'Title': title,
    'Model': model,
    'CC': CC,
    'Engine_type': Engine_type,
    'Transmission': Transmission,
    'Km_Driven': Km_Driven,
    'prices': prices
}

df = pd.DataFrame(Data)
df.to_csv(r"C:\Users\Lenovo\Desktop\All files\Pak Wheels Car Data2.csv")
```

```python
all_data = []

url = 'https://www.pakwheels.com/used-cars/search/-/?page=1'
response_car = requests.get(url)
soup_car = BeautifulSoup(response_car.text, 'html.parser')
script_tags = soup_car.find_all('script')
target_script = None
        
for script in script_tags:
    if 'application/ld+json' in script.get('type', ''):
        target_script = script
        break
if target_script:
    json_data = json.loads(target_script.string)
    if 'itemListElement' in json_data:
        items = json_data['itemListElement']
        for item in items:
            if "url" in item:
                all_data.append(item['url'])

data_rows = []

for url1 in all_data:
    response_car1 = requests.get(url1)
    soup_car1 = BeautifulSoup(response_car1.text, 'html.parser')
    ul = soup_car1.find('ul', {'class': 'list-unstyled ul-featured clearfix'})
    if ul:
        car_data = {}
        li_tags = ul.find_all('li', {'class': 'ad-data'})
        for li in li_tags:
            key = li.get_text(strip=True)
            value = li.find_next_sibling('li').get_text(strip=True)
            car_data[key] = value
        data_rows.append(car_data)

# Create DataFrame
df = pd.DataFrame(data_rows)

# Display DataFrame
df
```

```python
from selenium import webdriver
from bs4 import BeautifulSoup

url = 'https://www.pakwheels.com/used-cars/search/-/?page=1'

# Initialize a WebDriver (assuming you have ChromeDriver installed)
driver = webdriver.Firef()

# Load the URL
driver.get(url)

# Get the page source after JavaScript execution
page_source = driver.page_source

# Close the WebDriver
driver.quit()

# Parse the page source with BeautifulSoup
soup = BeautifulSoup(page_source, 'html.parser')

# Find the desired element
s = soup.find('ul', class_='price-box')
print(s)
```
